```python
#import library for manipulation
import pandas as pd
df=pd.DataFrame(data=titanic)
```

```python
#Show dataframe
df
```

```python
#create list of all column name
df_columns=df.columns
df_columns
```

```python
#lopp for print indiidual column for unique
for i in df_columns:
    print(i)
    print(df[i].unique())
```

```python
#age ,deck ,embarked,embarked_town hae nun value
df['embark_town']=df['embark_town'].replace('nan', None)
df['age']=df['age'].replace('nan', None)
```

```python
df
```

```python
#drop redundancy column
columns_to_drop = ['pclass', 'deck', 'embarked','deck', 'alive']
df=df.drop(columns=columns_to_drop)
```

```python
#add column 
df['ID'] = range(1, len(df) + 1)
df.insert(0, 'ID', df.pop('ID'))
df
```

```python
#change type for campatiable with sql
df['alone']=df['alone'].astype(str)
df['adult_male']=df['adult_male'].astype(str)
df['class']=df['class'].astype(str)
```
